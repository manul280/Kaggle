ã“ã‚Œã¯éå¸¸ã«é›£ã—ã„å•é¡Œã§ã™ã€‚ãƒ’ãƒ³ãƒˆã‚’ã‚‚ã‚‰ãˆã‚‹ã®ã§ãŠå¿˜ã‚Œãªã!

ãƒ«ã‚¤ãƒ¼ã‚¸ã¯ã€ãƒãƒªã‚ªã‚«ãƒ¼ãƒˆã‚µãƒ¼ã‚­ãƒƒãƒˆã§ãƒ¬ãƒ¼ã‚¹ã«å‹ã¤ãŸã‚ã®æœ€é©ãªã‚¢ã‚¤ãƒ†ãƒ ã‚’æ±ºå®šã™ã‚‹ãŸã‚ã«åˆ†æã‚’å®Ÿè¡Œã—ã‚ˆã†ã¨ã—ã¦ã„ã¾ã™ã€‚å½¼ã¯...ã®ã‚ˆã†ãªè¾æ›¸ã®ãƒªã‚¹ãƒˆã®å½¢ã§ã„ãã¤ã‹ã®ãƒ‡ãƒ¼ã‚¿ã‚’æŒã£ã¦ã„ã¾ã™ã€‚

    [
        {'name':'Peach'ã€ã€Œitems': ['ç·‘ã®ã‚·ã‚§ãƒ« ã€'banana'ã€'green shell'ã€]ã€'finish':3}ã€
        {'name':'Bowser'ã€ã€Œitems': ['ç·‘ã®ã‚·ã‚§ãƒ« ã€]ã€'finish':1}ã€
        #ãƒ¬ãƒ¼ã‚µãƒ¼ã®åå‰ãŒè¨˜éŒ²ã•ã‚Œã¦ã„ãªã„ã“ã¨ã‚‚ã‚ã‚‹
        {'name':ãªã—ã€ã€Œã‚¢ã‚¤ãƒ†ãƒ ': ['mushroom ã€]ã€'finish':2}ã€
        {'name':'Toad'ã€ã€Œitems': ['ç·‘ã®ã‚·ã‚§ãƒ« ã€'mushroom']ã€'finish':1}ã€
    ]

'items''ã¯ãƒ¬ãƒ¼ã‚µãƒ¼ãŒãã®ãƒ¬ãƒ¼ã‚¹ã§æ‹¾ã£ãŸã™ã¹ã¦ã®ãƒ‘ãƒ¯ãƒ¼ã‚¢ãƒƒãƒ—ã‚¢ã‚¤ãƒ†ãƒ ã®ãƒªã‚¹ãƒˆã§ã€''finish''ã¯ãƒ¬ãƒ¼ã‚¹ã§ã®é †ä½(1ä½ã¯1ã€3ä½ã¯3ãªã©)ã§ã‚ã‚‹ã€‚

å½¼ã¯ä»¥ä¸‹ã®é–¢æ•°ã‚’ä½œæˆã—ã¦ã€ã“ã®ã‚ˆã†ãªãƒªã‚¹ãƒˆã‚’å–å¾—ã—ã€å„é …ç›®ã‚’1ä½ã«ãªã£ãŸäººãŒé¸ã‚“ã å›æ•°ã«å¯¾å¿œã•ã›ãŸè¾æ›¸ã‚’è¿”ã—ã¾ã—ãŸã€‚

````python
def best_items(racers):
    """ãƒ¬ãƒ¼ã‚µãƒ¼è¾æ›¸ã®ãƒªã‚¹ãƒˆã‚’æŒ‡å®šã—ã¦ã€é …ç›®ã‚’1ä½ã§ã‚´ãƒ¼ãƒ«ã—ãŸãƒ¬ãƒ¼ã‚µãƒ¼ãŒãã‚Œã‚‰ã®é …ç›®ã‚’æ‹¾ã£ãŸå›æ•°ã«ãƒãƒƒãƒ—ã™ã‚‹è¾æ›¸ã‚’è¿”ã—ã¾ã™ã€‚
    """
    winner_item_counts = {}
    for i in range(len(racers)):
        # iç•ªç›®ã®ãƒ¬ãƒ¼ã‚µãƒ¼è¾æ›¸
        racer = racers[i]
        # 1ä½ã§ã‚´ãƒ¼ãƒ«ã—ãŸãƒ¬ãƒ¼ã‚µãƒ¼ã«ã—ã‹èˆˆå‘³ãŒãªã„
        if racer['finish'] == 1:
            for i in racer['items']:
                # ã“ã®é …ç›®ã®ã‚«ã‚¦ãƒ³ãƒˆã«1ã‚’è¿½åŠ ã—ã¾ã™ (å¿…è¦ã«å¿œã˜ã¦è¾æ›¸ã«è¿½åŠ ã—ã¾ã™) ã€‚
                if i not in winner_item_counts:
                    winner_item_counts[i] = 0
                winner_item_counts[i] += 1

        # ãƒ‡ãƒ¼ã‚¿å“è³ªã®å•é¡Œ:/åå‰ãŒè¨­å®šã•ã‚Œã¦ã„ãªã„ãƒ¬ãƒ¼ã‚µãƒ¼ã«é–¢ã™ã‚‹è­¦å‘Šã‚’å‡ºåŠ›ã—ã¾ã™ã€‚å¾Œã§å‡¦ç†ã—ã¾ã™ã€‚
        if racer['name'] is None:
            print("WARNING: Encountered racer with unknown name on iteration {}/{} (racer = {})".format(
                i+1, len(racers), racer['name'])
                 )
    return winner_item_counts
````
